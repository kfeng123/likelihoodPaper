```python
import numpy as np
from scipy.special import comb
from scipy.stats import binom
from scipy import optimize
import matplotlib.pyplot as plt
import math
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

n = 20
omega = 0.5
xi = 0.5
k = 4

def make_log_likelihood_ratio_function(obs):
    a = [0 for _ in range(k+1)]
    for i in obs:
        a[i] += 1

    def log_likelihood_ratio_function(omega, xi):
        out = 0
        for i in range(k + 1):
            tmp = a[i] * math.log( (1-omega) + 2**k * omega * (1-xi)**(k-i) * xi**i )
            out += tmp
        return out
    return log_likelihood_ratio_function


def log_LRT_stat(log_likelihood_ratio_function):
    def neg_log_likelihood_ratio_function(x):
        [omega, xi] = x
        return - log_likelihood_ratio_function(omega, xi)

    result = optimize.shgo(neg_log_likelihood_ratio_function, [(0, 1),(0+1e-6,1-1e-6)])
    result = optimize.shgo(lambda x:x[0]**2 + x[1]**2, [(0, 1),(0+1e-6,1-1e-6)])
    print(result)
    return - result.fun

def make_likelihood_ratio_function(obs):
    a = [0 for _ in range(k+1)]
    for i in obs:
        a[i] += 1

    def likelihood_ratio_function(x):
        [omega, xi] = x
        out = 1
        for i in range(k + 1):
            tmp = ((1-omega) + 2**k * omega * (1-xi)**(k-i) * xi**i ) ** a[i]
            out *= tmp
        return out
    return likelihood_ratio_function

def LRT_stat(likelihood_ratio_function):
    def neg_likelihood_ratio_function(x):
        [omega, xi] = x
        print(omega, xi)
        return - likelihood_ratio_function(omega, xi)


    result = optimize.shgo(neg_likelihood_ratio_function, [(0, 1),(0+1e-6,1-1e-6)])
    print(result)
    return - result.fun
```
